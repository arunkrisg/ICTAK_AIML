```python
# NumPy (Array Operations)
import numpy as np

# 1. Create a NumPy array of shape (5, 5) with random integers between 1 and 50

# NumPy array of shape (5, 5) with random integers between 1 and 50
arr = np.random.randint(1,51,size=(5,5))
print('\033[1m' + "The 5X5 array is given below: "+ '\033[0m')
print(arr)

# Taking sum of each row
# axis = 1 will conisder each rows of an array
# axis - 0 will consider each columns of an array
arr_sum = arr.sum(axis=1) # sum() and axis in Numpy will make the calculation of sum faster by overriding the use of loop
print('\033[1m' + "\nSum of each row in an array: "+ '\033[0m',arr_sum)

# This will find and display the maximum value in the array after summation of rows
max_sum = np.max(arr_sum)
print('\033[1m' + "\nThe maximum value: "+ '\033[0m',max_sum)

# Find the mean of the summation array
arr_mean = np.mean(arr_sum)
print('\033[1m' + "\nThe mean of the summation array: "+ '\033[0m',arr_mean)

# Replace all elements in summation array with 0 which is greater than the mean
arr_sum[arr_sum > arr_mean] = 0
print('\033[1m' + "\nFinal summation array after replacement: "+ '\033[0m',arr_sum)
```

```python
# 2. Create a NumPy array with numbers from 1 to 20

# NumPy array with numbers from 1 to 20
arry = np.array(range(1,21))
print('\033[1m' + "Original array: "+ '\033[0m',arry)

# Selecting every 3rd element starting from the 2nd element and assigning ot to a new array
new_arry = arry[1::3]
print('\033[1m' + "\nNew array: "+ '\033[0m',new_arry)
```

```python
# Pandas (Data Handling)
import pandas as pd
import seaborn as sns

# 3. Load the dataset Iris data

# Listing the different datasets from seaborn
print('\033[1m' + "Sample datasets in seaborn are:  \n"+ '\033[0m')
print(sns.get_dataset_names())

# Loading the Iris dataset from the lseaborn
ds = sns.load_dataset("iris")
print('\033[1m' + "\nThe Iris dataset is given below: \n"+ '\033[0m')
print(ds)

# Display the first 10 rows and basic statistics
print('\033[1m' + "\nThe first 10 rows from Iris dataset is given below: \n"+ '\033[0m')
print(ds.head(10))

# Display the summary of dataset
print('\033[1m' + "\nIris dataset summary: \n"+ '\033[0m')
print(ds.info())

# Display the statistical summary of dataset
print('\033[1m' + "\nIris dataset statistical summary: \n"+ '\033[0m')
print(ds.describe())

# Filter all rows where sepal_length > median and species is "versicolor"

# Find the median of sepal_length from Iris dataset
iris_median = ds["sepal_length"].median()
print('\033[1m' + "\nMedian of sepal_length is: "+ '\033[0m',iris_median)

# Filtering rows in Iris dataset
ds_fil = ds[(ds["sepal_length"] > iris_median) & (ds["species"] == "versicolor")]
print('\033[1m' + "\nFiltered Iris dataset is given below: \n"+ '\033[0m')
print(ds_fil)

# Group the data by species and finding the mean petal_length.
ds_grp = ds.groupby("species")["petal_length"].mean()
print('\033[1m' + "\nIris dataset after grouping and mean of petal_length is given below: \n"+ '\033[0m')
print(ds_grp)

# 4. Add a new column named petal_ratio = petal_length / petal_width
ds["petal_ratio"] = ds["petal_length"]/ds["petal_width"]
print('\033[1m' + "\nFinal Iris dataset after the addition of new column petal_ratio is given below: \n"+ '\033[0m')
print(ds)

# Find the top 5 rows with the highest petal_ratio

# Sorting the dataset in descending order based on petal_ratio
ds_sorted = ds.sort_values(by=["petal_ratio"],ascending=False)

# Selecting the top 5 rows
ds_t5 = ds_sorted.head(5)
print('\033[1m' + "\nTop 5 rows with the highest petal_ratio is given below: \n"+ '\033[0m')
print(ds_t5)
```

```python
# Data Visualization
import matplotlib.pyplot as plt

# 5. Using Matplotlib, create:

# A histogram of sepal_length with 15 bins
print('\033[1m' + "\nData Visualisation - Histogram\n"+ '\033[0m')
plt.hist(ds["sepal_length"], bins=15)
plt.xlabel("Sepal Length")
plt.ylabel("Frequency")
plt.title("Histogram of Sepal Length")
plt.show()

# A scatter plot of sepal_length vs petal_length, color-coded by species

# Separate data by species
setosa = ds[ds["species"] == "setosa"]
versicolor = ds[ds["species"] == "versicolor"]
virginica = ds[ds["species"] == "virginica"]

# Scatter Plot
print('\033[1m' + "\nData Visualisation - Scatterplot\n"+ '\033[0m')
plt.scatter(setosa["sepal_length"], setosa["petal_length"], label="Setosa")
plt.scatter(versicolor["sepal_length"], versicolor["petal_length"], label="Versicolor")
plt.scatter(virginica["sepal_length"], virginica["petal_length"], label="Virginica")

plt.xlabel("Sepal Length")
plt.ylabel("Petal Length")
plt.title("Sepal Length vs Petal Length")
plt.legend()
plt.show()

# 6.Using Seaborn, plot:

# A pairplot of the entire dataset

# Pairplot of Iris dataset
print('\033[1m' + "\nData Visualisation - Pairplot\n"+ '\033[0m')
sns.pairplot(ds, hue='species')
plt.show()

# Boxplot of petal_width by species
print('\033[1m' + "\nData Visualisation - Boxplot\n"+ '\033[0m')
sns.boxplot(x="species", y="petal_width", data=ds)
plt.xlabel("Species")
plt.ylabel("Petal Width")
plt.title("Petal Width Distribution by Species")
plt.show()
```
