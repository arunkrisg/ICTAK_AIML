```python
# Function to handle the outliers in the numerical columns of the new dataset
def handle_outliers_numeric(data, numeric_cols):
  for col in numeric_cols:
    Q1 = data[col].quantile(0.25)
    Q3 = data[col].quantile(0.75)

    IQR = Q3 - Q1

    LL = Q1 - 1.5 * IQR
    UL = Q3 + 1.5 * IQR

    # use .loc to modify safely
    # In the below code,
    # : - means all rows
    # col - only this column
    data.loc[:, col] = data[col].clip(LL,UL)

  return data
```

```python
# Display the newly created dataset
bnew_data.head(3)
```

```python
# Before go for encoding clean-up the data in 3 categorical columns in the dataset
# Here Engine, Max Power, and Max Torque should NOT be encoded as categorical variables
# These columns are numerical features stored as text
# Approach followed here is extracting numerical values from it by removing the text part
import numpy as np

# In the below converting the values in Max Torque to numbers
bnew_data["Max Torque"] = (
    bnew_data["Max Torque"]
    .astype(str)
    .str.extract(r"(\d+\.?\d*)")
    .astype(float)
)

# In the below converting the values in Engine and in Max Power to numbers
def clean_car_data(df):
    # Clean Engine column
    df['Engine'] = df['Engine'].astype(str)  # Ensure string type
    df['Engine'] = df['Engine'].str.replace(' cc', '', regex=False)  # Remove ' cc'
    df['Engine'] = pd.to_numeric(df['Engine'], errors='coerce')  # Convert to numeric, invalid -> NaN

    # Clean Max_Power column
    df['Max Power'] = df['Max Power'].astype(str)  # Ensure string type
    df['Max Power'] = df['Max Power'].str.split(' ').str[0]  # Take the number before 'bhp'
    df['Max Power'] = pd.to_numeric(df['Max Power'], errors='coerce')  # Convert to numeric

    return df

# Apply the cleaning
bnew_data = clean_car_data(bnew_data)

# Display the dataset after cleaning
bnew_data.head(5)
```

```python
# Checking the dataset again for missing values after data clean-up
# Calculate the percentage of null values
round(bnew_data.isnull().mean()*100,2)
```

```python
# There are missing values in Max Power column
# Fill missing Max Power with median
bnew_data['Max Power'] = bnew_data['Max Power'].fillna(bnew_data['Max Power'].median())
```

```python
# Checking the dataset again for missing values after filling the missing values
bnew_data.isnull().sum()

# The below result shows no more missing values
```

```python
# Skewness before transformation
print("Skewness of Max Power before transformation is: ",bnew_data["Max Power"].skew())

print("Skewness of Max Torque before transformation is: ",bnew_data["Max Torque"].skew())

print("Skewness of Engine before transformation is: ",bnew_data["Engine"].skew())
```

```python
# For Max Torque, the current skewness is between 0.5 and 1, we can use cube root transformation
bnew_data["Max Torque"] = np.cbrt(bnew_data["Max Torque"])

# After applying transformation, the skewness in the Column - Height
bnew_data["Max Torque"].skew()
```

```python
# For Engine, the current skewness is between 1 and 2, we can use cube root transformation
bnew_data["Engine"] = np.cbrt(bnew_data["Engine"])

# After applying transformation, the skewness in the Column - Height
bnew_data["Engine"].skew()
```

```python
# For Max Power, the current skewness is greater than 2, we can use log transformation
bnew_data["Max Power"] = np.log1p(bnew_data["Max Power"])

# After applying transformation, the skewness in the Column - Height
bnew_data["Max Power"].skew()
```

```python
# Skewness after transformation
print("Skewness of Max Power after transformation is: ",bnew_data["Max Power"].skew())

print("Skewness of Max Torque after transformation is: ",bnew_data["Max Torque"].skew())

print("Skewness of Engine after transformation is: ",bnew_data["Engine"].skew())
```

# **Machine Learning**

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression, Ridge, Lasso
# Score should be high and error should be low
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
```

```python
# This is for training a model
X_train, X_test, y_train, y_test = train_test_split(X_final,y,test_size=0.2,random_state=42)
```

### 1. Linear Regression

```python
lin_model = LinearRegression() # This is Multiple Linear Regression - multiple features and one target
lin_model.fit(X_train, y_train)
y_pred_in = lin_model.predict(X_test)

print("Linear Regression Metrics:")
print("R2:", r2_score(y_test,y_pred_in))
print("MAE:", mean_absolute_error(y_test,y_pred_in))
print("MSE:", mean_squared_error(y_test,y_pred_in))
print("RMSE:", np.sqrt(mean_squared_error(y_test,y_pred_in)))
```

### 2. Polynomial Regression

```python
poly = PolynomialFeatures(degree=2, include_bias=False)
X_train_poly = poly.fit_transform(X_train)
X_test_poly = poly.transform(X_test)

poly_model = LinearRegression()
poly_model.fit(X_train_poly, y_train)
y_pred_poly = poly_model.predict(X_test_poly)

print("Polynomial Regression Metrics:")
print("R2:", r2_score(y_test,y_pred_poly))
print("MAE:", mean_absolute_error(y_test,y_pred_poly))
print("MSE:", mean_squared_error(y_test,y_pred_poly))
print("RMSE:", np.sqrt(mean_squared_error(y_test,y_pred_poly)))
```

### 3. Ridge Regression

```python
ridge_model = Ridge(alpha=1.0)
ridge_model.fit(X_train, y_train)
y_pred_ridge = ridge_model.predict(X_test)

print("Ridge Regression Metrics:")
print("R2:", r2_score(y_test,y_pred_ridge))
print("MAE:", mean_absolute_error(y_test,y_pred_ridge))
print("MSE:", mean_squared_error(y_test,y_pred_ridge))
print("RMSE:", np.sqrt(mean_squared_error(y_test,y_pred_ridge)))
```

### 4. Lasso Regression

```python
lasso_model = Lasso(alpha=0.01)
lasso_model.fit(X_train, y_train)
y_pred_lasso = lasso_model.predict(X_test)

print("Ridge Regression Metrics:")
print("R2:", r2_score(y_test,y_pred_lasso))
print("MAE:", mean_absolute_error(y_test,y_pred_lasso))
print("MSE:", mean_squared_error(y_test,y_pred_lasso))
print("RMSE:", np.sqrt(mean_squared_error(y_test,y_pred_lasso)))
```

```python
# Create a dictionary to store metrics
metrics = {
    'Model':['Linear Regression','Polynomial Regression','Ridge Regression','Lasso Regression'],
    'R2'   :[
            r2_score(y_test,y_pred_in),
            r2_score(y_test,y_pred_poly),
            r2_score(y_test,y_pred_ridge),
            r2_score(y_test,y_pred_lasso)
    ],
    'MAE' :[
            mean_absolute_error(y_test,y_pred_in),
            mean_absolute_error(y_test,y_pred_poly),
            mean_absolute_error(y_test,y_pred_ridge),
            mean_absolute_error(y_test,y_pred_lasso)
    ],
    'MSE' :[
            mean_squared_error(y_test,y_pred_in),
            mean_squared_error(y_test,y_pred_poly),
            mean_squared_error(y_test,y_pred_ridge),
            mean_squared_error(y_test,y_pred_lasso)

    ],
    'RMSE':[
            np.sqrt(mean_squared_error(y_test,y_pred_in)),
            np.sqrt(mean_squared_error(y_test,y_pred_poly)),
            np.sqrt(mean_squared_error(y_test,y_pred_ridge)),
            np.sqrt(mean_squared_error(y_test,y_pred_lasso))
    ]

}

# Convert to Dataframe
final_data = pd.DataFrame(metrics)

# Display the comparison
print('\033[1m' + "The comparison between different models are given below: \n"+ '\033[0m')
final_data
```
