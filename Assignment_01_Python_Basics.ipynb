```python
# 1. Simple ATM Calculator

# Taking inputs from the user
balance = float(input("Enter the account balance: "))
withdrawal = float(input("Enter the withdrawal amount: "))

# Checking for sufficient balance
if withdrawal <= balance:
    balance -= withdrawal
    print(f"Transaction is successful. Remaining balance is: Rs {balance}")
else:
    print("Error: Insufficient balance.")
```

```python
# 2. Electricity Bill Generator

# Taking input from the user
units = float(input("Enter the number of units consumed: "))

# Initialize bill amount
bill = 0

# Calculating the bill based on slabs
if units <= 100:
    bill = units * 5
elif units <= 200:
    bill = 100 * 5 + (units - 100) * 7
else:
    bill = 100 * 5 + 100 * 7 + (units - 200) * 10

print(f"Bill amount = Rs {bill}")
```

```python
# 3. Temperature Converter

# Taking input from the user
temp = float(input("Enter the temperature value: "))
print("---------------")
print("C - Celsius")
print("F - Fahrenheit")
print("---------------")
unit = input("Enter the unit (C, F): ").upper()

# Conversion
if unit == 'C':
    # Celsius to Fahrenheit
    fahrenheit = (temp * 9/5) + 32
    print(f"Temperature in Fahrenheit: {fahrenheit}°F")
elif unit == 'F':
    # Fahrenheit to Celsius
    celsius = (temp - 32) * 5/9
    print(f"Temperature in Celsius: {celsius}°C")
else:
    print("Error: Invalid unit. Please enter 'C' or 'F'.")
```

```python
# 4. Mobile Recharge Validity

# datetime is the module and from there only importing the classes datetime & timedelta
from datetime import datetime, timedelta

# Taking input from the user
recharge_date_str = input("Enter recharge date (YYYY-MM-DD): ")
today_date_str = input("Enter today's date (YYYY-MM-DD): ")

# Converting string inputs to datetime objects
recharge_date = datetime.strptime(recharge_date_str, "%Y-%m-%d")
today_date = datetime.strptime(today_date_str, "%Y-%m-%d")

# Calculate expiry date (28 days from recharge)
expiry_date = recharge_date + timedelta(days=28) # This timedelta represents a duration, making it easier to perform arithmetic operations with time.
print(f"The Expiry date (YYYY-MM-DD): {expiry_date.date()}")

# Check validity
if today_date <= expiry_date:
    print("\nPlan is still valid.")
else:
    print("\nPlan expired. Recharge again.")
```

```python
# 5. Bus Seat Reservation System

# Initial list of passengers
booked = ['Amit', 'Sara', 'John']
print(f"Initial Passengers List: {booked}")

# Passenger to cancel
cancel = 'Sara'
if cancel in booked:
    booked.remove(cancel)  # Remove the cancelled passenger
print(f"Passenger list after cancelling Sara: {booked}")

# New passenger booking
new_passenger = 'Neha'
position = 1  # Position to insert (0-based index)

# Insert the new passenger at the specified position
booked.insert(position, new_passenger)

# Display the updated booking list
print(f"List after the addition of a new passenger: {booked}")
```

```python
# 6. Sum of even numbers

# Input from the user
num = int(input("Enter a number: "))

# Initialize the variable
sum_even = 0

# Loop through numbers from 1 to num
for i in range(1, num + 1):
    if i % 2 == 0:  # Check if the number is even
        sum_even += i

# Display the result
print(f"Sum of even numbers: {sum_even}")
```

```python
# 7. Dictionary – Student Marks Summary

# Initialize an empty dictionary
students = {}

# Loop to take inputs for 3 students
for i in range(3):
    name = input("Enter name: ")
    marks = float(input("Enter marks: "))
    students[name] = marks  # Store in dictionary
print(f"\nStudents name & marks: {students}")

# Calculate average marks
average = sum(students.values()) / len(students)

# Display the average
print(f"Average Marks: {average}")
```

```python
# 8. Dictionary – Word Frequency

# Input sentence from the user
sentence = input("Enter a sentence: ")

# Split the sentence into words
words = sentence.split()

# Initialize an empty dictionary
word_freq = {}

# Count the frequency of each word
for word in words:
    if word in word_freq:
        word_freq[word] += 1
    else:
        word_freq[word] = 1

# Display the word frequencies
for word, freq in word_freq.items():
    print(f"{word}: {freq}")
```

```python
# 10. Vowel Counter

# Function to count vowels in a string
def count_vowels(s):
    vowels = 'aeiouAEIOU'
    count = 0
    for char in s:
        if char in vowels:
            count += 1
    return count

# Take input from the user
text = input("Enter a string: ")

# Call the function and display result
num_vowels = count_vowels(text)
print(f"Number of vowels: {num_vowels}")
```

```python
# 11. Count Divisible Numbers

# Take input from the user
N = int(input("Enter the range limit (N): "))
D = int(input("Enter the divisor (D): "))

# Initialize an empty list to store divisible numbers
divisible_numbers = []

# Loop through numbers from 1 to N
for i in range(1, N + 1):
    if i % D == 0:
        divisible_numbers.append(i)

# Display the result
print(f"\nCount of numbers divisible by {D}: {len(divisible_numbers)}")
print(f"Numbers divisible by {D}: {divisible_numbers}")
```
